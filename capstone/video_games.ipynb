# Video Game Sales
Dataset obtained from [Kaggle](https://www.kaggle.com/gregorut/videogamesales), which in turn was scraped from [VGChartz](http://www.vgchartz.com/).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
games = pd.read_csv("vgsales.csv")
games.head()
```

```python
games.info()
```

```python
games.dropna().info()
```

```python
games.describe().transpose()
```

```python
plt.figure(figsize=(15, 11))

plt.subplot(2, 2, 1)
plt.hist(games["NA_Sales"], bins=50, color="#93cc49")
plt.title("Histogram: Video Game Sales (North America)")
plt.ylabel("Count")
plt.xlabel("Sales (in Millions of $)")

plt.subplot(2, 2, 2)
plt.hist(games["EU_Sales"], bins=50, color="#93cc49")
plt.title("Histogram: Video Game Sales (Europe)")
plt.ylabel("Count")
plt.xlabel("Sales (in Millions of $)")

plt.subplot(2, 2, 3)
plt.hist(games["JP_Sales"], bins=50, color="#93cc49")
plt.title("Histogram: Video Game Sales (Japan)")
plt.ylabel("Count")
plt.xlabel("Sales (in Millions of $)")

plt.subplot(2, 2, 4)
plt.hist(games["Other_Sales"], bins=50, color="#93cc49")
plt.title("Histogram: Video Game Sales (Other Regions)")
plt.ylabel("Count")
plt.xlabel("Sales (in Millions of $)")

plt.show()
```

```python
plt.figure(figsize=(7, 6))

plt.hist(games["Global_Sales"], bins=50, color="#93cc49")
plt.title("Histogram:Video Game Sales (Global)")
plt.ylabel("Count")
plt.xlabel("Sales (in Millions of $)")

plt.show()
```

```python
plt.figure(figsize=(7, 6))

plt.hist(np.log(games["Global_Sales"]), bins=50, color="#93cc49")
plt.title("Histogram: Log Video Game Sales (Global)")
plt.ylabel("Count")
plt.xlabel("Sales (in Millions of $)")

plt.show()
```

```python
plt.figure(figsize=(15, 11))
plt.boxplot([games["Global_Sales"], games["NA_Sales"], games["EU_Sales"], games["JP_Sales"], games["Other_Sales"]],
           labels=["Global", "North America", "Europe", "Japan", "Other Regions"])
plt.title("Video Game Sales Boxplots")
plt.ylabel("Sales in Millions of $")

plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.bar(games["Platform"].value_counts().index, games["Platform"].value_counts(), color="#42a4f4")

plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.bar(games["Genre"].value_counts().index, games["Genre"].value_counts(), color="#7d48cc")

plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.bar(games["Publisher"].value_counts().index, games["Publisher"].value_counts(), color="#93cc49")
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.bar(games["Publisher"].value_counts().head(50).index, games["Publisher"].value_counts().head(50), color="#93cc49")
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.plot(games["Year"].value_counts().sort_index())
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Number of Sales")

plt.show()
```
