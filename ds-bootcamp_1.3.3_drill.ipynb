```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
batting = pd.read_csv("mlb_team_batting_2008-2017.csv")
batting.head()
```

## 1. Choose one variable and plot that variable four different ways:

```python
ax = sns.distplot(batting["BA"], kde=False)
plt.title("Plot 1: Histogram for Batting Average (2008-2017)")
sns.despine()
ax.set(xlabel="BA", ylabel="")
plt.show()
```

```python
sns.distplot(batting["BA"], rug=True, kde=False)
plt.title("Plot 2: Histogram with Rugplot for Batting Average (2008-2017)")
sns.despine()
ax.set(xlabel="BA", ylabel="")
plt.show()
```

```python
sns.kdeplot(batting["BA"], shade=True)
plt.title("Plot 3: KDE Plot for Batting Average (2008-2017)")
sns.despine()
ax.set(xlabel="BA", ylabel="")
plt.show()
```

```python
sns.kdeplot(batting["BA"], shade=True, cut=0)
sns.rugplot(batting["BA"])
plt.title("Plot 4: KDE and Rugplot with Cut for Batting Average (2008-2017)")
sns.despine()
ax.set(xlabel="BA", ylabel="")
plt.show()
```

## 2. Choose two continuous variables, and plot them three different ways:

```python
g = sns.jointplot(x="BA", y="RunsPerGame", data=batting)
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Plot 1: Scatter Plot BA vs Teams Run Per Game (2008-2017)")
plt.show()
```

```python
g = sns.jointplot(x="BA", y="RunsPerGame", data=batting, kind="hex")
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Plot 2: Hex Plot BA vs Teams Run Per Game (2008-2017)")
plt.show()
```

```python
g = sns.jointplot(x="BA", y="RunsPerGame", data=batting, kind="kde")
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Plot 3: KDE Plot BA vs Teams Run Per Game (2008-2017)")
plt.show()
```

## 3. Choose one continuous variable and one categorical variable, and plot them six different ways:

```python
ax = sns.stripplot(x="lgID", y="BA", data=batting)
plt.title("Plot 1: Strip Plot - Batting Average by League (2008-2017)")
sns.despine(offset=10, trim=True)
ax.set(xlabel="", ylabel="BA")
plt.show()
```

```python
ax = sns.stripplot(x="lgID", y="BA", data=batting, jitter=True)
plt.title("Plot 2: Jitter Plot - Batting Average by League (2008-2017)")
sns.despine(offset=10, trim=True)
ax.set(xlabel="", ylabel="BA")

plt.show()
```

```python
ax = sns.swarmplot(x="lgID", y="BA", data=batting)
plt.title("Plot 3: Swarm Plot - Batting Average by League (2008-2017)")
sns.despine(offset=10, trim=True)
ax.set(xlabel="", ylabel="BA")

plt.show()
```

```python
ax = sns.boxplot(x="lgID", y="BA", data=batting)
plt.title("Plot 4: Box Plot - Batting Average by League (2008-2017)")
sns.despine(offset=10, trim=True)
ax.set(xlabel="", ylabel="BA")

plt.show()
```

```python
ax = sns.violinplot(x="lgID", y="BA", data=batting)
plt.title("Plot 5: Violin Plot - Batting Average by League (2008-2017)")
sns.despine(offset=10, trim=True)
ax.set(xlabel="", ylabel="BA")

plt.show()
```

```python
ax = sns.violinplot(x="lgID", y="BA", data=batting)
sns.swarmplot(x="lgID", y="BA", data=batting, color="black", alpha=0.5)
plt.title("Plot 6: Violin and Swarm Plot Batting Average by League (2008-2017)")
sns.despine(offset=10, trim=True)
ax.set(xlabel="", ylabel="BA")

plt.show()
```
